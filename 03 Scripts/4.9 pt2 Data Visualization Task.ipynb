# Import libraries

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

# Load data

```python
# define the path and assign dataframe to the "df" variable

path = r'C:\Users\Jakov\2023 Instacart Basket Analysis'

df = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'instacart_with_users.pkl'))
```

# 03.

```python
# plot a histogram with distribution of "order_hour_of_day" variable

histogram = df['order_hour_of_day'].plot.hist(bins = 24)
```

### The histogram tells that period when the vast majority orders are placed is from 09:00 till 16:00 with its peak between 10:00 and 11:00. From 17:00 until end of a day number of orders is gradually decreasing. A period with the least placed orders is between 00:00 and 07:00. 

# 04.

```python
# plot a pie chart with distribution of customers according to its loyalty

flags_count = df['loyalty_flag'].value_counts()
fig, ax = plt.subplots()
pie = ax.pie(flags_count,labels = flags_count.index, autopct='%1.0f%%')
```

# 05.

```python
# create a list with boolean values to map it later to a sample of the dataset

np.random.seed(5)
dev = np.random.rand(len(df)) <= 0.7
```

```python
# map the list with boolean values into sample of the dataset

big = df[dev]
small = df[~dev]
```

```python
# create a subset of the sampled dataset with neccessary columns only

df_small = small[['prices', 'order_hour_of_day']]
```

```python
# plot a line chart with columns from dataset sample

line_chart = sns.lineplot(data = df_small, x = 'order_hour_of_day',y = 'prices')
```

# 06.

```python
# create a subset of the sampled dataset with neccessary columns only

df_small_2 = small[['age', 'n_dependants']]
```

```python
# plot a line chart with columns from dataset sample

line_chart_2 = sns.lineplot(data = df_small_2, x = 'age', y = 'n_dependants')
```

# 07.

```python
# create a subset of the sampled dataset with neccessary columns only

df_small_3 = small[['age', 'income']]
```

```python
# plot a scatterplot with columns from the dataset sample

line_chart_3 = sns.scatterplot(data = df_small_3, x = 'age', y = 'income')
```
